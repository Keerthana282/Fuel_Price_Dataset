# **A Data-Driven Analysis of Fuel Price Trends and Volatility in Argentina**


## **Problem statment**
Fuel prices in Argentina vary across time, fuel types, and regions, creating uncertainty for consumers and businesses.
* Which fuel types are most expensive or volatile
* How prices change over time
* Whether certain regions consistently experience higher prices

This project aims to transform raw fuel price data into actionable insights that support informed decision-making.

# **Objective**

To analyze historical fuel price data in Argentina to identify pricing trends, regional variations, fuel-type differences, and volatility patterns using Python-based data analytics techniques.\

# **Project Outcomes**:
1 Clear identification of fuel price trends over time in Argentina.

2 Comparative insights into fuel prices across different fuel types

3 Identification of regions with consistently higher or lower fuel prices
Understanding of fuel price volatility and fluctuation patterns

4 Detection of periods with significant price increases or decreases

5 Improved understanding of economic impacts related to fuel price changes

```python
Q1=df['Price (Arg Pesos)'].quantile(0.25)
Q3=df['Price (Arg Pesos)'].quantile(0.75)
IQR=Q3-Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df=df[(df['Price (Arg Pesos)'] >= lower_bound) &
    (df['Price (Arg Pesos)'] <= upper_bound)]

df
```

```python
sns.boxplot(df[(df['Price (Arg Pesos)'] >= lower_bound) &
    (df['Price (Arg Pesos)'] <= upper_bound)])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(y=o2)
plt.title("Before Outlier Removal")

plt.subplot(1,2,2)
sns.boxplot(y=df['Price (Arg Pesos)'])
plt.title("After Outlier Removal")

plt.tight_layout()
plt.show()
```

```python
year_avg = df.groupby('Year & Month ID')['Price (Arg Pesos)'].mean()

plt.figure(figsize=(10,6))
plt.bar(year_avg.index, year_avg.values)
plt.xlabel("Year & Month")
plt.ylabel("Average Fuel Price")
plt.title("Average Fuel Price by Year")
plt.xticks(rotation=90)
plt.show()
```

```python
product = df.groupby('Product Type')['Price (Arg Pesos)'].sum()

plt.figure(figsize=(8,5))
plt.bar(product.index, product.values)
plt.xlabel("Product Type")
plt.ylabel("Price")
plt.title("Fuel Price by Product Type")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date_unix'] = df['Date'].astype('int64') // 10**9
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price (Arg Pesos)'],marker='v',color='r',linestyle=':')
plt.xlabel("Date")
plt.ylabel("Fuel Price (Arg Pesos)")
plt.title("Fuel Price Trend Over Time in Argentina")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

*Key Insights: Fuel Price Trend in Argentina*

1)Strong upward trend over time
Fuel prices show a clear and consistent increase across the timeline, indicating long-term inflationary pressure in the fuel market.

2)Acceleration in recent periods
The rise toward the right side of the chart suggests that fuel prices increased much faster in recent years compared to earlier periods.

3)High price volatility
The scattered and dense red points, especially in later dates, indicate frequent price fluctuations, reflecting economic instability and frequent price revisions.

4)Widening price range over time
Earlier prices are tightly clustered, while later prices span a much wider range (roughly from ~140 to over 550 ARS), showing increased uncertainty and market stress.


```python
product = df.groupby('Product Type')['Price (Arg Pesos)'].mean()
```

```python
plt.figure(figsize=(8,5))
plt.bar(product.index, product.values)
plt.xlabel("Product Type")
plt.ylabel("Average Fuel Price (Arg Pesos)")
plt.title("Average Fuel Price by Product Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

*Key Insights*:

1)GASOIL GRADE 3 is the most expensive fuel type, indicating higher refining quality or performance usage.

2)PETROL PREMIUM â 95 OCTANE is consistently priced higher than regular unleaded fuel.

3)COMPRESSED NATURAL GAS (CNG) is the cheapest option, making it a cost-efficient alternative for consumers.

4)GASOIL GRADE 2 and PETROL UNLEADED (95/98 OCTANE) occupy the mid-price segment.

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Region',y='Price (Arg Pesos)',hue='Product Type',estimator='mean')

plt.title('Average Fuel Price by Region and Product Type')
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Key Insights:

1)NEA and CENTRO regions show the highest fuel prices across most product types, indicating stronger demand or higher logistics costs.

2)PATAGONIA generally records lower average prices, possibly due to regional subsidies or lower taxation.

3)Price ranking among fuel types remains consistent across regions (Grade 3 > Premium > Grade 2 > Unleaded > CNG).

4)Regional price differences are more pronounced for premium fuels, suggesting uneven cost pass-through.

5)CNG prices vary the least across regions, indicating more stable pricing controls.

```python
plt.figure(figsize=(9,5))
sns.countplot(data=df,x='Region',hue='Product Type')
plt.title('Product Type by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
Region_avg = df.groupby('Region')['Price (Arg Pesos)'].mean()

plt.figure(figsize=(12,6))
plt.bar(Region_avg.index, Region_avg.values)
plt.xlabel("Region")
plt.ylabel("Average Fuel Price (Arg Pesos)")
plt.title("Average Fuel Price by Region")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="Shift Type",y="Price (Arg Pesos)",hue="Shift Type")
plt.xlabel("Shift Type")
plt.ylabel("Fuel Price (Arg Pesos)")
plt.title("Fuel Price Comparison by Shift Type")
plt.tight_layout()
plt.show()
```

```python
product_counts = df['Product Type'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Fuel Distribution by Product Type")
plt.show()
```

```python
pivot_table = df.pivot_table(values='Price (Arg Pesos)',index='Product Type',columns='Shift Type',aggfunc='mean')
plt.figure(figsize=(8,6))
sns.heatmap(pivot_table,annot=True,cmap='YlOrRd',fmt='.1f')
plt.title("Average Fuel Price by Product Type and Shift Type")
plt.tight_layout()
plt.show()
```
